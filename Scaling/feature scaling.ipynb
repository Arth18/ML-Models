# Feature Scaling

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score , root_mean_squared_error 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler , RobustScaler
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Feature Scaling Resource16994544271.txt')
```

```python
data
```

```python
scaler = StandardScaler()
min_max_scaler = MinMaxScaler()
robust_scaler = RobustScaler()
```

```python
data[['P1','P2']] = scaler.fit_transform(data[['Pregnancies','Glucose']])   #STD Scaling
data[['B1','S1']] = min_max_scaler.fit_transform(data[['BloodPressure','SkinThickness']])   #Min-Max Scaling
data[['I1','BMI2']] = robust_scaler.fit_transform(data[['Insulin','BMI']])  #Robust Scaling
```

```python
data.head(4)
```

```python
sns.kdeplot(data['P1'])
sns.kdeplot(data['P2'])
plt.title('Standard Scaling')
```

```python
sns.kdeplot(data['B1'],color='blue')
sns.kdeplot(data['S1'],color='red')
plt.title('Min-Max Scaling')
```

```python
sns.kdeplot(data['I1'],color='green')
sns.kdeplot(data['BMI2'],color='orange')
plt.title('Robust Scaling')
```
