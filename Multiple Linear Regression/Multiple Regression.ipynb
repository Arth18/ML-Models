```python
from sklearn.linear_model import Ridge , Lasso
from sklearn.metrics import r2_score , root_mean_squared_error 
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("ml 2 SLR.txt")
```

```python
data.head(2)
```

```python
model1 = Ridge(alpha=1.0)
model2 = Lasso(alpha=0.1)
```

```python
x = data.drop("Sell Price", axis = 1)
y = data['Sell Price']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
model1.fit(x_train , y_train)
model2.fit(x_train , y_train)
y_pred1 = model1.predict(x_test)
#y_pred2 = model2.predict(x_test)
```

```python
y_pred2 = model2.predict(x_test)
```

```python
r2_score(y_test , y_pred1)*100
```

```python
r2_score(y_test , y_pred2)*100
```

```python
root_mean_squared_error(y_test , y_pred1)
```

```python
root_mean_squared_error(y_test , y_pred2)
```
