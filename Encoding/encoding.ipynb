# LABEL ENCODING 

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
```

```python
Data = {"Type": ["None", "LOW", "MEDIUM", "HIGH", "VERY HIGH", "None", "LOW", "MEDIUM", "HIGH"]}
data = pd.DataFrame(Data)
```

```python
encoder = LabelEncoder()
data["Label_Encoded"] = encoder.fit_transform(data["Type"])
```

# One HOT ENCODING

```python
a = pd.get_dummies(data["Type"])
```
